## Extracción de Datos a Nivel Nacional

```python
import requests

import pandas as pd

pd.set_option('display.max_columns', None) 
```

```python
#url 2011

url = "https://apidatos.ree.es/es/datos/generacion/estructura-generacion?start_date=2011-01-01T00:00&end_date=2011-12-31T23:59&time_trunc=month"

energia = requests.get(url)

if energia.status_code != 200:
    print(f"Error. El motivo es {energia.reason}")
else:
    print(f"La respuesta es {energia.status_code}")
    generacion_energia = energia.json()
```

```python
#estabamos cogiendo la fecha de donde no era
generacion_energia["included"][0]["attributes"]["values"][0]["datetime"]
```

```python
#ejemplo solo 2011
dicc_generacion = {"tipo_energia":[], "tipo_generacion":[], "fecha":[], "valor":[]}

for gen in generacion_energia["included"]:
    for e in gen["attributes"]["values"]:
        dicc_generacion ["tipo_energia"].append(gen["attributes"]['title'])
        dicc_generacion ["tipo_generacion"].append(gen["attributes"]['type'])
        dicc_generacion ["fecha"].append(e["datetime"])
        dicc_generacion ["valor"].append(e['value'])
    
```

```python
df_generacion = pd.DataFrame(dicc_generacion)
df_generacion
```

```python
#de 2011 a 2022, url dinamica, creamos lista para guardar las url

lista_url = []

for año in range(2011, 2023):

    url = f"https://apidatos.ree.es/es/datos/generacion/estructura-generacion?start_date={año}-01-01T00:00&end_date={año}-12-31T23:59&time_trunc=month"
    
    lista_url.append(url)
lista_url
```

```python
#creamos lista para guardar los json de cada año, cogemos urls de la lista anterior

lista_json = []
for url in lista_url:

    energia = requests.get(url)

    if energia.status_code != 200:
        print(f"Error. El motivo es {energia.reason}")
    else:
        print(f"La respuesta es {energia.status_code}")
        generacion_energia = energia.json()
        lista_json.append(generacion_energia)
```

```python
dicc_generacion = {"tipo_energia":[], "tipo_generacion":[], "fecha":[], "valor":[]}

for json in lista_json: #entramos en cada json de lista jsons
    for gen in json["included"]: #entramos en la key included que es donde estan los datos
        for e in gen["attributes"]["values"]: #entramos en atributos luego values que es donde estan las fechas y los valores
            #al hacer todos los append en este for, los campos que son iguales para x fechas se repetiran 
            #tantas veces como sea necesario, asi no dara el error de que los arrays no son igual de largos
            dicc_generacion ["tipo_energia"].append(gen["attributes"]['title']) 
            dicc_generacion ["tipo_generacion"].append(gen["attributes"]['type'])
            dicc_generacion ["fecha"].append(e["datetime"])
            dicc_generacion ["valor"].append(e['value'])
```

```python
#dataframe desde 2011 a 2022 dividido por mes
df_generacion_total = pd.DataFrame(dicc_generacion)
df_generacion_total
```

## Extracción de Datos por Comunidad Autónoma

```python
#diccionario codigos ccaa
cod_comunidades = {'Ceuta': 8744,
                    'Melilla': 8745,
                    'Andalucía': 4,
                    'Aragón': 5,
                    'Cantabria': 6,
                    'Castilla - La Mancha': 7,
                    'Castilla y León': 8,
                    'Cataluña': 9,
                    'País Vasco': 10,
                    'Principado de Asturias': 11,
                    'Comunidad de Madrid': 13,
                    'Comunidad Foral de Navarra': 14,
                    'Comunitat Valenciana': 15,
                    'Extremadura': 16,
                    'Galicia': 17,
                    'Illes Balears': 8743,
                    'Canarias': 8742,
                    'Región de Murcia': 21,
                    'La Rioja': 20}
```

```python
dicc_generacion_ccaa = {"tipo_energia":[], "tipo_generacion":[], "fecha":[], "valor":[], "ccaa":[]}
#añadimos el campo para comunidades a la url
for k,v in cod_comunidades.items():
    for año in range(2011, 2023):

        url = f"https://apidatos.ree.es/es/datos/generacion/estructura-generacion?start_date={año}-01-01T00:00&end_date={año}-12-31T23:59&time_trunc=month&geo_limit=ccaa&geo_ids={v}"

        energia = requests.get(url)

        if energia.status_code != 200:
            print(f"Error. El motivo es {energia.reason}")
        else:
            #print(f"La respuesta es {energia.status_code}")
            generacion_energia = energia.json()

        
        for gen in generacion_energia["included"]: #entramos en la key included que es donde estan los datos
            for e in gen["attributes"]["values"]: #entramos en atributos luego values que es donde estan las fechas y los valores
            #al hacer todos los append en este for, los campos que son iguales para x fechas se repetiran 
            #tantas veces como sea necesario, asi no dara el error de que los arrays no son igual de largos
                dicc_generacion_ccaa ["tipo_energia"].append(gen["attributes"]['title']) 
                dicc_generacion_ccaa ["tipo_generacion"].append(gen["attributes"]['type'])
                dicc_generacion_ccaa ["fecha"].append(e["datetime"])
                dicc_generacion_ccaa ["valor"].append(e['value'])
                dicc_generacion_ccaa ["ccaa"].append(k)
            
dicc_generacion_ccaa
```

```python
#dataframe desde 2011 a 2022 dividido por mes
df_generacion_total_ccaa = pd.DataFrame(dicc_generacion_ccaa)
df_generacion_total_ccaa
```
